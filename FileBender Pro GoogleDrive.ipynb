```
#@title Zip file (Make ZIP archive)
import os
from datetime import datetime
now = datetime.now()
date_now = now.strftime("%D - %H:%M:%S")

directory = '/content/drive/My Drive/GalleryDownloaded/pixiv/' #@param {type:"string"}
source = '8252709 hara0742' #@param {type:"string"}
zipfile = '8252709 hara0742.zip' #@param {type:"string"}
quiet_mode = True #@param {type:"boolean"}

print("\33[32m"+"Runtime started at "+date_now+"\33[0m")
print('\33[93m'+'Initializing..'+'\33[0m')
%cd "$directory"
print('\33[34m'+'Compressing..'+'\33[0m')

if quiet_mode == True:
  !zip -P "kuronekony4n" -r -q "$zipfile" "$source"
else:
  !zip -P "kuronekony4n" -r "$zipfile" "$source"

print('\33[34m'+'Archive completed.'+'\33[0m')

file_stats = os.stat(directory+"/"+zipfile)
print('Saved as > '+'\33[93m'+zipfile+'\33[0m'+f' ({file_stats.st_size / (1024 * 1024)}MB)')

now = datetime.now()
date_now = now.strftime("%D - %H:%M:%S")
print("\33[32m"+"Runtime ended at "+date_now+"\33[0m")
```

```
#@title Unzip file (Extract ZIP)
from datetime import datetime
now = datetime.now()
date_now = now.strftime("%D - %H:%M:%S")

directory = '/content/drive/My Drive/DownloadFromURL/' #@param {type:"string"}
zipfile = 'Negligee UI.zip' #@param {type:"string"}
extract_to = 'Negligee UI' #@param {type:"string"}
quiet_mode = True #@param {type:"boolean"}

print("\33[32m"+"Runtime started at "+date_now+"\33[0m")
print('\33[93m'+'Initializing..'+'\33[0m')

%cd "$directory"
print("Making directory with name'",extract_to,"'")
%mkdir "$extract_to"
print("Directory created.")
print('\33[34m'+'Extracting..'+'\33[0m')

if quiet_mode == True:
  !unzip -q "$zipfile" -d "$extract_to"
else:
  !unzip "$zipfile" -d "$extract_to"

print('\33[34m'+'Archive completed.'+'\33[0m')
print("Extracted to > "+'\33[93m'+extract_to+'\33[0m')

now = datetime.now()
date_now = now.strftime("%D - %H:%M:%S")
print("\33[32m"+"Runtime ended at "+date_now+"\33[0m")
```

```
#@title Copy file/directory
from subprocess import call
from datetime import datetime
now = datetime.now()
date_now = now.strftime("%D - %H:%M:%S")

source = '/content/drive/My Drive/#Source/[KissSub][Steins;Gate \uFF06 Steins;Gate0][BDrip][GB-BIG5-JP][X265-10bit][1080P][HEVC]/\u53D1\u5E03\u9875.txt' #@param {type:"string"}
target = '/content/drive/My Drive/#Source/Steins;Gate Copied' #@param {type:"string"}

def cp_dir(source, target):
    call(['cp', '-a', source, target]) 

print("\33[32m"+"Runtime started at "+date_now+"\33[0m")
print('\33[93m'+'Initializing..'+'\33[0m')
print('\33[34m'+'Copying..'+'\33[0m')

cp_dir(source, target)

print('\33[34m'+'Copy Complete, but make sure your file is safely copied..'+'\33[0m')

now = datetime.now()
date_now = now.strftime("%D - %H:%M:%S")
print("\33[32m"+"Runtime ended at "+date_now+"\33[0m")
```
